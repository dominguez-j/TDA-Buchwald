# 1 - Elemento desordenado
Implementar, por división y conquista, una función que dado un arreglo sin elementos repetidos y casi ordenado (todos los elementos se encuentran ordenados, salvo uno), obtenga el elemento fuera de lugar.

```python
def elemento_desordenado(arr):
    return elemento_desordenado_rec(arr, 0, len(arr) -1)

def elemento_desordenado_rec(arr, izq, der):
    if der - izq == 1:
        if arr[izq] > arr[der]:
            return arr[izq]
        return -1

    mid = (izq + der)//2

    if arr[mid -1] > arr[mid]:
        return arr[mid -1]

    mitad_der = elemento_desordenado_rec(arr, mid, der)
    if mitad_der == -1:
        return elemento_desordenado_rec(arr, izq, mid)
    return mitad_der
```

# 2 - Unos y ceros

Se tiene un arreglo tal que [1, 1, 1, …, 0, 0, …] (es decir, unos seguidos de ceros). Se pide una función de orden O(log(n)) que encuentre el índice del primer 0. 
Si no hay ningún 0 (solo hay unos), debe devolver -1.

```python
def indice_primer_cero(arr):
    return indice_primer_cero_rec(arr, 0, len(arr) -1)

def indice_primer_cero_rec(arr, izq, der):
    if izq > der: return -1

    mid = (izq + der) // 2

    if arr[mid] == 0:
        mitad_izq = indice_primer_cero_rec(arr, izq, mid -1)
        if mitad_izq == -1:
            return mid
        else:
            return mitad_izq
    return indice_primer_cero_rec(arr, mid +1, der)
```

# 3 - Raiz cuadrada

Implementar un algoritmo que, por división y conquista, permita obtener la parte entera de la raíz cuadrada de un número n, en tiempo O(log n). 
Por ejemplo, para n = 10 debe devolver 3, y para n = 25 debe devolver 5.

```python
def parte_entera_raiz(n):
    return parte_entera_rec(n, 0, n)

def parte_entera_rec(n, izq, der):
    if izq > der: return der
    
    mid = (izq + der) //2
    cuadrado = mid * mid

    if cuadrado == n:
        return mid
    elif cuadrado < n:
        return parte_entera_rec(n, mid +1, der)

    return parte_entera_rec(n, izq, mid -1)
```

# 4 - Picos
Se tiene un arreglo de N >= 3 elementos en forma de pico, esto es: estrictamente creciente hasta una determinada posición p, 
y estrictamente decreciente a partir de ella (con 0 < p < N - 1). 
Por ejemplo, en el arreglo [1, 2, 3, 1, 0, -2] la posición del pico es p = 2. Se pide:

Implementar un algoritmo de división y conquista de orden O(log n) que encuentre la posición p del pico: 
func PosicionPico(v []int, ini, fin int) int. 
La función será invocada inicialmente como: PosicionPico(v, 0, len(v)-1), y tiene como pre-condición que el arreglo tenga forma de pico.

```python
def posicion_pico(v, ini, fin):
    mid = (ini + fin)//2

    if v[mid -1] < v[mid] and v[mid] > v[mid +1]:
        return mid
    elif v[mid -1] > v[mid]:
        return posicion_pico(v, ini, mid)
    return posicion_pico(v, mid, fin)
```

# 5 - Merge sort

```python
def merge_sort(arr):
    if len(arr) <= 1: return arr

    mid = len(arr) // 2

    izq = merge_sort(arr[:mid])
    der = merge_sort(arr[mid:])

    return merge(izq, der)

def merge(izq, der):
    resultado = []
    i = j = 0

    # Compara los elementos de ambas listas y agrégalos en orden
    while i < len(izq) and j < len(der):
        if izq[i] <= der[j]:
            resultado.append(izq[i])
            i += 1
        else:
            resultado.append(der[j])
            j += 1

    # Agrega los elementos restantes de la lista izquierda, si hay
    while i < len(izq):
        resultado.append(izq[i])
        i += 1

    # Agrega los elementos restantes de la lista derecha, si hay
    while j < len(der):
        resultado.append(der[j])
        j += 1

    return resultado
```

# 6 - Multiplicacion de numeros grandes
Implementar un algoritmo de multiplicación de dos números grandes de longitud n, por división y conquista, 
con un orden de complejidad mejor que O(n^2).

```python
def multiplicar(a, b):

    if a < 10 or b < 10:
        return a * b

    n = max(len(str(a)), len(str(b)))
    m = n // 2

    a1 = a // 10**m
    a0 = a % 10**m

    b1 = b // 10**m
    b0 = b % 10**m

    c0 = multiplicar(a0, b0)
    c1 = multiplicar(a1,b1)
    p = multiplicar((a1 + a0),(b1 + b0))

    return (c1 * 10**(2*m)) + ((p -c1 -c0) * 10**m) + c0
```

# 9 - Más de la mitad de repeticiones
Implementar una función (que utilice división y conquista) de orden O(n logn) que dado un arreglo de n números enteros devuelva true o false según si existe algún elemento que aparezca más de la mitad de las veces.

### Ejemplos:

[1, 2, 1, 2, 3] -> false

[1, 1, 2, 3] -> false

[1, 2, 3, 1, 1, 1] -> true

[1] -> true

```python
def conteo_dyc(arr, izq, der):

    if izq >= der:
        return arr[izq]

    mid = (izq + der)//2

    candidato_izq = conteo_dyc(arr, izq, mid)
    candidato_der = conteo_dyc(arr, mid +1, der)
    
    count_izq = count_der = 0
    for i in range(izq,der +1):
        if candidato_izq == arr[i]:
            count_izq += 1
        if candidato_der == arr[i]:
            count_der += 1
            
    if count_izq > (der - izq +1)//2:
        return candidato_izq
    if count_der > (der - izq +1)//2:
        return candidato_der
    return None

def mas_de_la_mitad(arr):
    return conteo_dyc(arr, 0, len(arr) -1)
```

# 10 - Más de la mitad en O(n)
Implementar una función (que utilice división y conquista) de orden O(n) que dado un arreglo de n números enteros devuelva true o false según si existe algún elemento que aparezca más de la mitad de las veces.

```python
def mas_de_la_mitad(arr):

    if len(arr) == 0:
        return None
    elif len(arr) == 1:
        return arr[0]

    reduccion = []

    for i in range(0, len(arr) -1, 2):
        if arr[i] == arr[i +1]:
            reduccion.append(arr[i])

    candidato = mas_de_la_mitad(reduccion)

    if candidato is not None and arr.count(candidato) > len(arr) // 2:
        return candidato
    if len(arr) % 2 != 0 and arr.count(arr[-1]) > len(arr) // 2:
        return arr[-1]
    return None
```

# 11 - Más de dos tercios de O(n)
Implementar una función (que utilice división y conquista) de orden O(n) que dado un arreglo de n números enteros devuelva true o false según si existe algún elemento que aparezca más de dos tercios de las veces.

```python
def mas_de_dos_tercios(arr):
    
    if len(arr) == 0:
        return None
    elif len(arr) == 1:
        return arr[0]

    reduccion = []

    for i in range(0, len(arr) -1, 2):
        if arr[i] == arr[i +1]:
            reduccion.append(arr[i])

    candidato = mas_de_dos_tercios(reduccion)

    if candidato is not None and arr.count(candidato) > len(arr) * 2 // 3:
        return candidato
    fif len(arr) % 2 != 0 and arr.count(arr[-1]) > len(arr) * 2 // 3:
        return arr[-1]
    return None
```

# 13 - Subarreglo de máxima suma
Dado un arreglo de n enteros (no olvidar que pueden haber números negativos), encontrar el subarreglo contiguo de máxima suma, utilizando División y Conquista.

### Ejemplos

[5, 3, 2, 4, -1] ->  [5, 3, 2, 4]

[5, 3, -5, 4, -1] ->  [5, 3]

[5, -4, 2, 4, -1] -> [5, -4, 2, 4]

[5, -4, 2, 4] -> [5, -4, 2, 4]

[-3, 4, -1, 2, 1, -5] -> [4, -1, 2, 1]

```python
def max_subarray(arr):
    n = len(arr)
    if n <= 1:
        return arr
    
    ini = 0
    fin = n - 1
    
    i_arr, f_arr, suma = max_subarray_dyc(arr, ini, fin)
    return arr[i_arr:f_arr + 1] 

def sumar_medio(arr, ini, mid, fin):

    sum_izq = 0
    sum_max_izq = float('-inf')

    max_izq = mid

    for i in range(mid, ini - 1, -1):
        sum_izq += arr[i]
        if sum_izq > sum_max_izq:
            sum_max_izq = sum_izq
            max_izq = i
    
    sum_der = 0
    sum_max_der = float('-inf')

    max_der = mid

    for i in range(mid + 1, fin + 1):
        sum_der += arr[i]
        if sum_der > sum_max_der:
            sum_max_der = sum_der
            max_der = i
    
    return max_izq, max_der, sum_max_izq + sum_max_der

def max_subarray_dyc(arr, ini, fin):

    if ini == fin:
        return ini, fin, arr[ini]

    mid = (ini + fin) // 2

    i_izq, f_izq, sum_izq = max_subarray_dyc(arr, ini, mid)
    i_der, f_der, sum_der = max_subarray_dyc(arr, mid + 1, fin)
    i_medio, f_medio, sum_mid = sumar_medio(arr, ini, mid, fin)

    if sum_izq >= sum_der and sum_izq >= sum_mid:
        return i_izq, f_izq, sum_izq
    elif sum_der >= sum_izq and sum_der >= sum_mid:
        return i_der, f_der, sum_der
    else:
        return i_medio, f_medio, sum_mid
```

# 15 - Bolzano
Se sabe, por el teorema de Bolzano, que si una función es continua en un intervalo [a, b], y que en el punto a es positiva y en el punto b es negativa (o viceversa), necesariamente debe haber (al menos) una raíz en dicho intervalo. Implementar una función raiz que reciba una función (univariable) y los extremos mencionados a y b, y devuelva una raíz dentro de dicho intervalo (si hay más de una, simplemente quedarse con una). La complejidad de dicha función debe ser logarítmica del largo del intervalo [a, b]. Asumir que por más que se esté trabajando con números enteros, hay raíz en dichos valores: Se puede trabajar con floats, y el algoritmo será equivalente, simplemente se plantea con ints para no generar confusiones con la complejidad.

```python
def raiz(funcion, a, b):
    if funcion(a) * funcion(b) > 0:
        return None

    if funcion(a) * funcion(b) == 0:
        if funcion(a) == 0: 
            return a
        return b

    m = (a + b) // 2

    if funcion(m) == 0:
        return m

    if funcion(a) * funcion(m) < 0: 
        return raiz(funcion, a, m)
    return raiz(funcion, m, b)
```
